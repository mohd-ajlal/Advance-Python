```python
import os
from os import path
```

```python
print(os.name)
```

```python
print(path.exists("text.txt"))
print(path.isfile("text.txt"))
print(path.isdir("folder"))
```

```python
print(path.realpath("text.txt"))
print(path.split(path.realpath("text.txt")))
```

```python
import time

print(time.ctime(path.getmtime("text.txt")))
```

```python
from datetime import datetime

print(datetime.fromtimestamp(path.getmtime("text.txt")))
```

```python
td = datetime.now() - datetime.fromtimestamp(path.getmtime("text.txt"))
print(td)
print(td.total_seconds())
```
