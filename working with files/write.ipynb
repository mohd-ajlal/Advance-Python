```python
# open a file for writiing and create it if it does not exist

sample_file = open("text.txt", "w+")
sample_file.write("This is a sample file.\n")
sample_file.close()
```

```python
# Open the file for appending text to the end of the file

sample_file = open("text.txt", "a")
sample_file.write("This is another line.\n")
sample_file.close()
```

```python
# write some line of data to the file

with open("text.txt", "w") as sample_file:
    for i in range(5):
        sample_file.write(f"This is line {i}\n")
        
```
