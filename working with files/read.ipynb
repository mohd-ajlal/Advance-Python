```python
# Open the file and read the contents

with open("text.txt", "r") as sample_file:
    contents = sample_file.read()
    print(contents)
```

```python
samplefile = open("text.txt", "r")
if samplefile.mode == 'r':
     contents = samplefile.read()
     print(contents)
samplefile.close()
```

```python
# readlines() returns a list of lines in the file

with open("text.txt", "r") as sample_file:
    lines = sample_file.readlines()
    print(lines)
```
