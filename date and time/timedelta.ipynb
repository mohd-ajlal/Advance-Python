```python
# Timedelta

# Timedelta represents a duration, the difference between two dates or times. It is part of the datetime module in Python.
from datetime import datetime, timedelta, date
```

```python
# contructing a timedelta object
delta = timedelta(days=365, hours=3, minutes=30)
print(delta)  # Output: 365 days, 3:30:00
```

```python
now = datetime.now()
print("Current date and time:", now)
```

```python
# print todays days oen year from now

print("One year from now:", now + timedelta(days=365))
```

```python
# create a deltatime that uses more than one argument

print("Delta:", now + timedelta(days=2, hours=6, minutes=15))
```

```python
# calculate one week ago from now, format it in string format

print("One week ago:", (now - timedelta(weeks=1)).strftime("%Y-%m-%d %H:%M:%S"))
```

```python
# How many days until april fools day?

today = date.today()
april_fools = date(today.year, 4, 1)
if today > april_fools:
    april_fools = date(today.year + 1, 4, 1)
days_until_april_fools = (april_fools - today).days
print(f"Days until April Fools' Day: {days_until_april_fools}")
```

```python

today = date.today()  # get today's date
afd = date(today.year, 4, 1)  # get April Fool's for the same year
```

```python
# use date comparison to see if April Fool's has already gone for this year
# if it has, use the replace() function to get the date for next year
# today = datetime.datetime(2024, 12, 9)
if afd < today:
    print (f"April Fool's day already went by {(today-afd).days} days ago")
    afd = afd.replace(year=today.year + 1)  # if so, get the date for next year

# Now calculate the amount of time until April Fool's Day  
time_to_afd = afd - today
print (f"It's just {time_to_afd.days} days until next April Fools' Day!")
```
