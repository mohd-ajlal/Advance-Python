```python
from datetime import datetime
from datetime import date
```

```python
today = date.today()
print("Today's date:", today)
```

```python
print("Date Components:", today.day, "-", today.month, "-", today.year)
```

```python
# retrieving today weekday
print("Today's Weekday:", today.weekday())
# 0: Monday, 1: Tuesday, 2: Wednesday, 3: Thursday, 4: Friday, 5: Saturday, 6: Sunday
```

```python
# get today's date + time with datetime class

today_datetime = datetime.today()
print("Today's date and time:", today_datetime)
```

```python
today_date_time = datetime.now()    
print("Today's date and time:", today_date_time)
```

```python
# get today's date with datetime class

today_date = datetime.today().date()
print("Today's date:", today_date)
```

```python
# Get current time
current_time = datetime.now().time()
print("Current time:", current_time)
```

```python
from datetime import time

# Create a time object
my_time = time(14, 30, 45)  # 2:30:45 PM
print("My time:", my_time)
```
