```python
import math
```

```python
# Constants
print(math.pi)
print(math.e)

print(math.nan)
print(math.inf)
print(-math.inf)
```

```python
# Trigonometry
obst_direction = math.cos(math.pi / 4)
print(obst_direction)
print(math.sin(math.pi / 4))
```

```python
# Ceiling and Floor
cookies = 10.3
candy = 7
print(math.ceil(cookies))
print(math.ceil(candy))

age = 47.9
otherAge = 47
print(math.floor(age))
print(math.floor(otherAge))
```
