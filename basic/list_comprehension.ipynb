```python
myList = [1,2,3,4,5]
[x**2 for x in myList]
```

```python
# List comprehension is a concise way to create lists. It consists of brackets containing an expression followed by a for clause, then zero or more for or if clauses. The expressions can be anything, meaning you can put in all kinds of objects in lists. The result will be a new list resulting from evaluating the expression in the context of the for and if clauses which follow it.
```

```python
# Syntax:
# [expression for item in iterable if condition == True]
```

```python
# List comprehension with filters

myList = [1,2,3,4,5]
[x**2 for x in myList if x % 2 == 0]
```

```python
# List comprehension with functions

def square(x):
    return x**2
myList = [1,2,3,4,5]
[square(x) for x in myList]
```

```python
# Nested list comprehension

matrix = [[1,2,3],[4,5,6],[7,8,9]]
[[row[i] for row in matrix] for i in range(3)]
```
