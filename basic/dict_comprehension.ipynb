# Dictionay Comprehension

## What is Dictionay Comprehension



```python
# Dict comprehension -> https://www.programiz.com/python-programming/dictionary-comprehension
```

```python
dict_comprehension = {x: x**2 for x in range(5)}
print(dict_comprehension)
```

```python
animal_Dict = {'cat': 4, 'dog': 4, 'spider': 8}
squared_Dict = {key: value**2 for key, value in animal_Dict.items()}
print(squared_Dict)
```

```python
# Syntax for dict comprehension:
# {key: value for item in iterable if condition == True}
```
