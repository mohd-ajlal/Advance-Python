```python
# Dictionary 
```

```python
myDict = {'name': 'Ajlal', 'age': 22, 'city': 'Basti'}
print(myDict)
```

```python
myd = {'name': 'Ajlal',
    'age': 22, 
    'city': 'Basti', 
    'hobbies': ['coding', 'gaming', 'traveling'],
    'percentage': 85.5
    }

print(myd)
```

```python
print(myd['name']) # This will print 'Ajlal'
print(myd['hobbies']) # This will print the list of hobbies
```

```python
myd['seven'] = 7
print(myd)
```

```python
print('seven' in myd) # This will return True because 'seven' is a key in myd.
print('fullName' in myd) # This will return False because 'fullName' is not a key in myd.
```

```python
print(myd.keys()) # This will print all the keys in the dictionary.
print(myd.values()) # This will print all the values in the dictionary.
```

```python
for key, value in myd.items():
    print(f"{key}: {value}")
```
